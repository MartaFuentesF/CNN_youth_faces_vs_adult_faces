### Visualizing Some of our data from each class

```python
# Function to make a list of 'num_images' images
    # from a specific folder

def load_images_from_class(class_folder, num_images):
    class_images = [] # images will be appended here
    
    path_class_folder = f'/Users/marta/Documents/image_data/train/{class_folder}/'
    class_files = os.listdir(path_class_folder)
    random.shuffle(class_files)
    
    # iterate through the files and keep only 'num_images' number of files
    for i, file in enumerate(class_files):
        if i == num_images:
            break
            
        img_path = os.path.join(path_class_folder, file)
        img = Image.open(img_path)
        class_images.append(img)
        
    return class_images
```

```python
# citation: Hank Buttler
def image_samples(img_list, num_images):
    temp = load_images_from_class(img_list, num_images)
    for i in temp:
        plt.clf()
        plt.imshow(i)
        plt.show()
   
```

```python
image_samples('class_1', 5)
image_samples('class_2', 5)
```
